# Data Wrangling

To prepare the analysis part, the datasets need to be reshaped, joined, cleaned. Here, I will reshape the dataset first, and then join the datasets. After that, the null values will be removed.  

## Modify datasets

In this part, I will modify (reshape) and join the dataset, so that every attributes in a same dataframe. After joining and modifying the datsets, columns of the processed dataset might look like: {Country, Year, Indicator 1 Value, Indicator 2 Value, … }

```python
def modify_dataframe(df, column_name):
    '''
    This function is for reshaping a dataset.
    After reshaping, the columns in the dataframe will look like {Country, Year, Indicator}
    ------
    Input:
        df (pandas.DataFrame): a dataframe which will be modified
        column_name
    
    '''
    target_ = df.copy().set_index('country').stack(dropna=False).reset_index()
    target_ = target_.rename(columns={'level_1':'year',
                                      0:column_name    
    })    
    return target_
```

```python
# modify the income dataframe
df_income = modify_dataframe(df_income,'income_per_person')
df_income.head()
```

```python
# apply same process to other datasets
df_co2 = modify_dataframe(df_co2, 'co2_per_person')
df_life = modify_dataframe(df_life, 'life_expectancy')
df_child = modify_dataframe(df_child, 'child_mortality')
```

```python
# join the datasets
df = df_income.merge(df_co2, on=['country','year'], how='outer')
df = df.merge(df_life, on=['country','year'], how='outer')
df = df.merge(df_child, on=['country','year'], how='outer')
```

# Exploratory Data Analysis

## Q1) How have indicators changed in global scale?

Personal income and life expectancy increased from 1970 to 2018. The life expectancy graph shows a linearly increasing trend in that period, however, the graph of the personal income is more fluctuating than the graph of life expectancy. <br><br>
Otherwise, CO2 emission and child mortality decreased from 1970 to 2018. The line chart of child mortality shows a linearly decreasing trend in the same period. However, the change of CO2 emission largely fluctuated in that period. <br><br>
When I checked the correlation between 'year', 'personal income', 'life expectancy', and 'child mortality', the personal income and life expectancy have a strong and positive correlation with 'year', and the child mortality was strongly and negatively correlated with 'year'.

```python
sns.set_style('whitegrid')
plt.figure()
plt.subplots(2,2,figsize=(18,7))
plt.subplot(2,2,1)
plt.title('Income per person')
sns.lineplot(data=df_cleaned.groupby('year').mean()['income_per_person'])
plt.subplot(2,2,2)
plt.title('CO2 emission per person')
sns.lineplot(data=df_cleaned.groupby('year').mean()['co2_per_person'])
plt.subplot(2,2,3)
plt.title('Life expectancy')
sns.lineplot(data=df_cleaned.groupby('year').mean()['life_expectancy'])
plt.subplot(2,2,4)
plt.title('Child mortality')
sns.lineplot(data=df_cleaned.groupby('year').mean()['child_mortality'])
```

```python
plt.figure(figsize=(9,7))
plt.title('Correlation between the variables in global scale')
sns.heatmap(df_cleaned.groupby('year',as_index=False).mean().corr(), annot=True, cmap='YlGnBu')
```

## How were the other indicators correlated with personal income level?

```python
income_median = df_cleaned['income_per_person'].median()
```

```python
df_high_income = df_cleaned.loc[df_cleaned['income_per_person']>income_median]
df_low_income = df_cleaned.loc[df_cleaned['income_per_person']<=income_median]
```

```python
df_high_income.shape, df_low_income.shape
```

```python
df_cleaned.corr()
```

### CO2 Emission (tonnes per person)

According to the below histograms, the group of higher-income shows bigger CO2 emissions than the group of lower-income than the median. In addition to this, a scatter plot below shows CO2 emission and personal income have a moderate positive correlation regarding the local scale. When we think about the CO2 emission has a negative correlation with 'year' and the personal income has a positive correlation with 'year' regarding the global scale, the result looks opposite to the result of the local scale.

```python
plt.figure(figsize=(15,7))
plt.title('Histogram of CO2 emissions (tonnes per person)', fontsize=14)
sns.distplot(df_high_income['co2_per_person'], color='#F44336', hist=True, kde=False, label='income > median')
sns.distplot(df_low_income['co2_per_person'], color='#2196F3', hist=True, kde=False, label='income <= median')
plt.xlabel('CO2 per person', fontsize=12)
plt.legend(fontsize=12)
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(x=df_cleaned['income_per_person'], y=df_cleaned['co2_per_person'])
plt.xlabel('income per person (GDP/capita, PPP$ inflation-adjusted)',fontsize=12)
plt.ylabel('CO2 Emission per person', fontsize=12)
plt.title('CO2 Emission per person VS income per person', fontsize=14)
```

## Life expectancy

Like CO2 emission, life expectancy in the higher-income group is mostly higher than the lower-income group's life expectancy. However, the correlation between life expectancy and personal income is different by income level. With the group has a personal income of less than 50000, the correlation is positive. However, when we check the group has a personal income of more than 50000, the correlation is negative.

```python
plt.figure(figsize=(15,7))
plt.title('Histogram of Life expectancy', fontsize=14)
sns.distplot(df_high_income['life_expectancy'], color='#F44336', hist=True, kde=False, label='income > median' )
sns.distplot(df_low_income['life_expectancy'], color='#2196F3', hist=True, kde=False, label='income <= median')
plt.xlabel('Life Expectancy', fontsize=12)
plt.legend(fontsize=12)
```

```python
plt.figure(figsize=(15,7))
plt.axvline(x=50000, color='red', linestyle='dashed')
sns.scatterplot(x=df_cleaned['income_per_person'], y=df_cleaned['life_expectancy'])
plt.xlabel('income per person (GDP/capita, PPP$ inflation-adjusted)', fontsize=12)
plt.ylabel('Life Expectancy', fontsize=12)
plt.title('Life Expectancy VS income per person', fontsize=14)
```

## Child Mortality

Unlikely to CO2 emission rate and life expectancy, most of the child mortality of the higher-income group is lower than that of lower-income group. According to below scatter plot, the correlation with child mortality and personal income is negative at the income level of less than 100000, however, in the group has income of more than 100000, the correlation is positive.

```python
plt.figure(figsize=(15,7))
plt.title('Histogram of Child Mortality', fontsize=14)
sns.distplot(df_high_income['child_mortality'], color='#F44336', hist=True, kde=False, label='income > median')
sns.distplot(df_low_income['child_mortality'], color='#2196F3', hist=True, kde=False, label='income <= median')
plt.xlabel('Child Mortality', fontsize=12)
plt.legend(fontsize=12)
```

```python
plt.figure(figsize=(15,7))
plt.axvline(x=100000, color='red', linestyle='dashed', label='income = 100000')
sns.scatterplot(x=df_cleaned['income_per_person'], y=df_cleaned['child_mortality'])
plt.xlabel('Income per person (GDP/capita, PPP$ inflation-adjusted)', fontsize=12)
plt.ylabel('Child Mortality', fontsize=12)
plt.title('Child Mortality VS income per person', fontsize=14)
```
